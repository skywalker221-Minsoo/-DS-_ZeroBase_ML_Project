- Premium Tech Support: 고객이 이 회사에서 제공하는 기다림 시간이 감소된 추가 기술 지원 계획을 구독하는지 여부 (예, 아니오)
- Streaming TV: 고객이 인터넷 서비스를 사용하여 타사 공급업체로부터 텔레비전 프로그램을 스트리밍하는지 여부 (예, 아니오). 회사는 이 서비스에 추가 비용을 부과하지 않습니다.
- Streaming Movies: 고객이 인터넷 서비스를 사용하여 타사 공급업체로부터 영화를 스트리밍하는지 여부 (예, 아니오). 회사는 이 서비스에 추가 비용을 부과하지 않습니다.

- <span style="color: black; font-size: medium; background-color: #fff5b1;">Churn: 이탈 여부 (이탈: Yes, 이탈하지 않음: No)</span>

#### AgeGroup 생성 

```python
age_bins = [10, 19, 29, 49, 59, 90]
age_labels = ['Teen', 'YoungAdult', 'PrimeAdult','MiddleAdult', 'Senior']
churn_data['AgeGroup'] = pd.cut(churn_data['Age'], bins=age_bins, labels=age_labels)
churn_data[['Age', 'AgeGroup']] 
```

```python
plt.figure(figsize=(12,6))
colors = sns.color_palette('magma', n_colors=4)
ax = churn_data.AgeGroup.value_counts().reindex(age_labels).plot.barh(color=colors)

# - Teen : 19세
# - YoungAdult : 20세 ~ 29세
# - PrimeAdult :  30세 ~ 49세
# - MiddleAdult : 50세 이상 ~ 60세
# - Senior : 60세 이상 

age_group = ['Teen (19세 이하)', 'YoungAdult (20세 - 29세)','PrimeAdult (30세 - 49세)', 'MiddleAdult (50세 - 59세)', 'Senior (60세 이상)']
# y 축 텍스트 변경
ax.set_yticklabels(age_group, size = 12)

for i, v in enumerate(churn_data['AgeGroup'].value_counts().reindex(age_labels)):
    if i==0:
        plt.text(v, i, str(v), color='black', size = 12, weight = 'bold')
    else:
        plt.text(v-150, i, str(v), color='white', size = 12, weight = 'bold')

ax.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Count', fontsize=12)
plt.gca().invert_yaxis() 
plt.show()
```

```python
# 나이 범주에 따른 이탈 퍼센트 생성 
age_group_df = pd.crosstab(churn_data.AgeGroup, churn_data.Churn, values=churn_data.Churn,aggfunc='count')
counts = age_group_df.sum(axis=1)
age_group_df['PctNo'] = round(age_group_df['No'] / counts * 100)
age_group_df['PctYes'] = round(age_group_df['Yes'] / counts * 100)
age_group_df
```

```python
age_group_df[['PctNo', 'PctYes']].sort_index().plot(
    figsize=(12, 6),
    kind='barh',
    color=['#CED7EA', '#ff0f4f'],
    stacked=True
)
    
for i, v in enumerate(age_group_df['PctNo'].sort_index()):
    plt.text(v+(1-v)/2 , i, f"{int(v)}% ({age_group_df['No'][i]})", color='black', size = 12, weight = 'bold')



for i, v in enumerate(age_group_df['PctYes'].sort_index()):
    plt.text((100-v)+(v/3.5) , i, f"{int(v)}% ({age_group_df['Yes'][i]})", color='white', size = 12, weight = 'bold')

plt.legend(title='Churn', labels=['No', 'Yes'])
plt.title('Churn Percent of Age Group',fontsize=15, weight='bold' )
plt.show()
```
