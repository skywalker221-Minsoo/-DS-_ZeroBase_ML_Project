#### Column 정보 


#### Demographics 

- <span style="color: black; font-size: medium; background-color: #fff5b1;">CustomerID: 각 고객을 식별하는 고유 ID</span>
- Count: 필터링된 집합의 고객 수를 합산하는 데 사용되는 값
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Gender: 고객의 성별 (남성, 여성) </span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> Age: 고객의 현재 연령 (분기의 끝에서의 연령) </span>
- Senior Citizen: 고객이 65세 이상인지 여부 (예, 아니오)
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> Married: 고객이 결혼했는지 여부 (예, 아니오) </span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> Dependents: 고객이 부양 가족과 함께 살고 있는지 여부 (예, 아니오). 부양 가족은 자녀, 부모, 조부모 등이 될 수 있습니다. </span>
- Number of Dependents: 고객과 함께 살고 있는 부양 가족의 수

#### Location

-<span style="color: black; font-size: medium; background-color: #fff5b1;"> Country: 고객의 주요 거주 국가 : United States 로 통일 <span> 
-<span style="color: black; font-size: medium; background-color: #fff5b1;"> State: 고객의 주요 거주 주(State) : California로 통일 </span>
-<span style="color: black; font-size: medium; background-color: #fff5b1;"> City: 고객의 주요 거주 도시</span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> Zip Code: 고객의 주요 거주 우편 번호
- Lat Long: 고객의 주요 거주지의 위도와 경도의 결합
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> Latitude: 고객의 주요 거주지의 위도 </span>
-<span style="color: black; font-size: medium; background-color: #fff5b1;"> Longitude: 고객의 주요 거주지의 경도 </span>

#### Services

- Referred a Friend: 고객이 이 회사를 친구나 가족에게 추천했는지 여부 (예, 아니오)
- Number of Referrals: 고객이 지금까지 추천한 참조 수
- Tenure in Months: 해당 분기의 끝까지 회사에 머무른 총 월 수
- Offer: 고객이 마지막으로 수락한 마케팅 제안 (해당되는 경우). None, Offer A, Offer B, Offer C, Offer D, Offer E와 같은 값으로 구성됩니다.
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Phone Service: 고객이 이 회사와 가정용 전화 서비스를 구독하는지 여부 (예, 아니오) </span>
- Avg Monthly Long Distance Charges: 고객의 평균 장거리 통화 요금 (분기의 끝까지 계산)
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Multiple Lines: 고객이 이 회사와 여러 전화 라인을 구독하는지 여부 (예, 아니오) </span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Internet Service: 고객이 이 회사와 인터넷 서비스를 구독하는지 여부 (아니오, DSL, Fiber Optic, Cable) </span>
- Avg Monthly GB Download: 고객의 평균 다운로드 용량(GB) (분기의 끝까지 계산)
- Online Security: 고객이 이 회사에서 제공하는 추가 온라인 보안 서비스를 구독하는지 여부 (예, 아니오)
- Online Backup: 고객이 이 회사에서 제공하는 추가 온라인 백업 서비스를 구독하는지 여부 (예, 아니오)
- Device Protection Plan: 고객이 이 회사에서 제공하는 인터넷 장비의 추가 장치 보호 계획을 구독하는지 여부 (예, 아니오)
- Premium Tech Support: 고객이 이 회사에서 제공하는 기다림 시간이 감소된 추가 기술 지원 계획을 구독하는지 여부 (예, 아니오)
- Streaming TV: 고객이 인터넷 서비스를 사용하여 타사 공급업체로부터 텔레비전 프로그램을 스트리밍하는지 여부 (예, 아니오). 회사는 이 서비스에 추가 비용을 부과하지 않습니다.
- Streaming Movies: 고객이 인터넷 서비스를 사용하여 타사 공급업체로부터 영화를 스트리밍하는지 여부 (예, 아니오). 회사는 이 서비스에 추가 비용을 부과하지 않습니다.
- Streaming Music: 고객이 인터넷 서비스를 사용하여 타사 공급업체로부터 음악을 스트리밍하는지 여부 (예, 아니오). 회사는 이 서비스에 추가 비용을 부과하지 않습니다.
- Unlimited Data: 고객이 무제한 데이터 다운로드/업로드를 위해 월별 추가 요금을 지불했는지 여부 (예, 아니오)
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Contract: 고객의 현재 계약 유형 (월간, 1년, 2년) </span>


#### Charge

- Paperless Billing: 고객이 전자 청구서를 선택했는지 여부 (예, 아니오)
- Payment Method: 고객이 청구서를 결제하는 방법 (은행 인출, 신용 카드, 우편으로 수표)
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Monthly Charge: 고객의 현재 서비스별 월간 총 요금</span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Total Charges: 고객의 총 청구액 (분기의 끝까지 계산)</span>
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Total Refunds: 고객의 총 환불액 (분기의 끝까지 계산) </span>
- Total Extra Data Charges: 고객의 플랜에 명시된 데이터 이외의 추가 데이터 다운로드에 대한 총 요금 (분기의 끝까지 계산)
- Total Long Distance Charges: 고객의 플랜에 명시된 장거리 통화 이외의 총 요금 (분기의 끝까지 계산)

#### Status

- <span style="color: black; font-size: medium; background-color: #fff5b1;">Satisfaction Score: 고객의 회사에 대한 전반적인 만족도 평가 (1-매우 불만족, 5-매우 만족)</span>
- Satisfaction Score Label: 점수 (1-5)의 텍스트 버전을 나타냄
- <span style="color: black; font-size: medium; background-color: #fff5b1;">Customer: 분기말에 고객의 상태를 나타냄 (이탈, 유지, 가입)</span>
- Churn Label: 이탈 여부 (이탈: Yes, 이탈하지 않음: No)
- Churn Score: IBM SPSS Modeler를 사용하여 계산된 0-100 사이의 값. 이탈을 유발하는 여러 요소를 포함한 모델입니다. 점수가 높을수록 고객의 이탈 가능성이 높습니다.
- Churn Score Category: 다음 범주 중 하나에 Churn Score를 할당하는 계산: 0-10, 11-20, 21-30, 31-40, 41-50, 51-60, 61-70, 71-80, 81-90 및 91-100
- <span style="color: black; font-size: medium; background-color: #fff5b1;"> CLTV: 고객 생애 가치 (CLTV). 기업의 공식과 기존 데이터를 사용하여 예측된 CLTV입니다. 값이 높을수록 고객의 가치가 높습니다.</span>
- CLTV Category: 다음 범주 중 하나에 CLTV 값을 할당하는 계산: 2000-2500, 2501-3000, 3001-3500, 3501-4000, 4001-4500, 4501-5000, 5001-5500, 5501-6000, 6001-6500 및 6501-7000.
- Churn Reason : 이탈 이유

#### AgeGroup 생성 

```python
age_bins = [10, 19, 29, 49, 59, 90]
age_labels = ['Teen', 'YoungAdult', 'PrimeAdult','MiddleAdult', 'Senior']
churn_data['AgeGroup'] = pd.cut(churn_data['Age'], bins=age_bins, labels=age_labels)
churn_data[['Age', 'AgeGroup']] 
```

```python
plt.figure(figsize=(12,6))
colors = sns.color_palette('magma', n_colors=4)
ax = churn_data.AgeGroup.value_counts().reindex(age_labels).plot.barh(color=colors)

# - Teen : 19세
# - YoungAdult : 20세 ~ 29세
# - PrimeAdult :  30세 ~ 49세
# - MiddleAdult : 50세 이상 ~ 60세
# - Senior : 60세 이상 

age_group = ['Teen (19세 이하)', 'YoungAdult (20세 - 29세)','PrimeAdult (30세 - 49세)', 'MiddleAdult (50세 - 59세)', 'Senior (60세 이상)']
# y 축 텍스트 변경
ax.set_yticklabels(age_group, size = 12)

for i, v in enumerate(churn_data['AgeGroup'].value_counts().reindex(age_labels)):
    if i==0:
        plt.text(v, i, str(v), color='black', size = 12, weight = 'bold')
    else:
        plt.text(v/2-40, i, str(v), color='white', size = 12, weight = 'bold')

ax.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Count', fontsize=12)
plt.gca().invert_yaxis() 
plt.show()
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.countplot(data=churn_data, x='AgeGroup', hue='Churn', palette=['#E9738C', '#89AAE3'])

plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Churn Count by Age Group', fontsize=15, weight='bold')

plt.show()
```

```python
age_group_df = pd.crosstab(churn_data.AgeGroup, churn_data.Churn, values=churn_data.Churn,aggfunc='count')
counts = age_group_df.sum(axis=1)
age_group_df['PctNo'] = round(age_group_df['No'] / counts * 100)
age_group_df['PctYes'] = round(age_group_df['Yes'] / counts * 100)
```

```python
age_group_df[['PctNo', 'PctYes']].sort_index().plot(
    figsize=(12, 6),
    kind='barh',
    color=['#CED7EA', '#ff0f4f'],
    stacked=True
)
    
for i, v in enumerate(age_group_df['PctNo'].sort_index()):
    plt.text(v+(1-v)/2 , i, f"{int(v)}%", color='black', size = 14, weight = 'bold')

for i, v in enumerate(age_group_df['PctYes'].sort_index()):
    plt.text((100-v)+(v/2.5) , i, f"{int(v)}%", color='white', size = 14, weight = 'bold')

plt.legend(title='Churn', labels=['No', 'Yes'])
plt.title('Churn Percent of Age Group',fontsize=15, weight='bold' )
plt.show()
```
