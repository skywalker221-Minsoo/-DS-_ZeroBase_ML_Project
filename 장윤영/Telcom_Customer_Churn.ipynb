#### Status

- <span style="color: black; font-size: medium; background-color: #fff5b1;">Satisfaction Score: 고객의 회사에 대한 전반적인 만족도 평가 (1-매우 불만족, 5-매우 만족)</span>

#### AgeGroup 생성 

```python
age_bins = [10, 19, 29, 49, 59, 90]
age_labels = ['Teen', 'YoungAdult', 'PrimeAdult','MiddleAdult', 'Senior']
churn_data['AgeGroup'] = pd.cut(churn_data['Age'], bins=age_bins, labels=age_labels)
churn_data[['Age', 'AgeGroup']] 
```

```python
plt.figure(figsize=(12,6))
colors = sns.color_palette('magma', n_colors=4)
ax = churn_data.AgeGroup.value_counts().reindex(age_labels).plot.barh(color=colors)

# - Teen : 19세
# - YoungAdult : 20세 ~ 29세
# - PrimeAdult :  30세 ~ 49세
# - MiddleAdult : 50세 이상 ~ 60세
# - Senior : 60세 이상 

age_group = ['Teen (19세 이하)', 'YoungAdult (20세 - 29세)','PrimeAdult (30세 - 49세)', 'MiddleAdult (50세 - 59세)', 'Senior (60세 이상)']
# y 축 텍스트 변경
ax.set_yticklabels(age_group, size = 12)

for i, v in enumerate(churn_data['AgeGroup'].value_counts().reindex(age_labels)):
    if i==0:
        plt.text(v, i, str(v), color='black', size = 12, weight = 'bold')
    else:
        plt.text(v-150, i, str(v), color='white', size = 12, weight = 'bold')

ax.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Count', fontsize=12)
plt.gca().invert_yaxis() 
plt.show()
```

```python
# 나이 범주에 따른 이탈 퍼센트 생성 
age_group_df = pd.crosstab(churn_data.AgeGroup, churn_data.Churn, values=churn_data.Churn,aggfunc='count')
counts = age_group_df.sum(axis=1)
age_group_df['PctNo'] = round(age_group_df['No'] / counts * 100)
age_group_df['PctYes'] = round(age_group_df['Yes'] / counts * 100)
age_group_df
```

```python
age_group_df[['PctNo', 'PctYes']].sort_index().plot(
    figsize=(12, 6),
    kind='barh',
    color=['#CED7EA', '#ff0f4f'],
    stacked=True
)
    
for i, v in enumerate(age_group_df['PctNo'].sort_index()):
    plt.text(v+(1-v)/2 , i, f"{int(v)}% ({age_group_df['No'][i]})", color='black', size = 12, weight = 'bold')



for i, v in enumerate(age_group_df['PctYes'].sort_index()):
    plt.text((100-v)+(v/3.5) , i, f"{int(v)}% ({age_group_df['Yes'][i]})", color='white', size = 12, weight = 'bold')

plt.legend(title='Churn', labels=['No', 'Yes'])
plt.title('Churn Percent of Age Group',fontsize=15, weight='bold' )
plt.show()
```

### 범주형 데이터를 숫자형 데이터로 변경 
- 모델 학습을 위해 범주형 데이터를 숫자형 데이터로 변경한다.
- Yes, No 로 구분되는 데이터를 Yes = 1, No = 0 으로 변경한다 
- Contract 컬럼의 계약 유형도 개월수 데이터로 변경해준다 - Month-to-month : 1, One year : 12, Two year : 24, 
- gender 컬럼은 남자는 1 여자는 0 으로 변경 


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

# gender 
le.fit((churn_data['gender'])) 
churn_data['gender_n'] = le.transform(churn_data['gender']) # Male = 1, Female = 0
# Married
le.fit((churn_data['Married'])) 
churn_data['Married_n'] = le.transform(churn_data['Married']) # No = 0, Yes = 1
# PhoneService
le.fit((churn_data['PhoneService'])) 
churn_data['PhoneService_n'] = le.transform(churn_data['PhoneService']) # Yes = 1, No = 0
# MultipleLines
le.fit((churn_data['MultipleLines'])) 
churn_data['MultipleLines_n'] = le.transform(churn_data['MultipleLines']) # Yes = 1, No = 0
```

```python
# Contract
contract_type = churn_data['Contract'].unique()
# 계약 유형을 개월수로 변경 
churn_data['ContractMonth'] = churn_data['Contract'].apply(lambda x: 1 if x == contract_type[0] else (24 if x == contract_type[1] else 12))

# Churn
le.fit((churn_data['Churn'])) 
churn_data['Churn_n'] = le.transform(churn_data['Churn']) # Yes = 1, No = 0

churn_data.head()
```
